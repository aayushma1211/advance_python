## Question 10: Reduce Function
Use `functools.reduce` to find the maximum number in a list without using the built-in `max()` function.
```python
numbers = [55, 12, 89, 34, 72]
```

````python
# Write your answer here
# reduce() is a function from the functools module.
# It reduces a list (or iterable) to a single value by applying a function cumulatively.
# Syntax:
# reduce(function, iterable)

from functools import reduce

numbers = [55, 12, 89, 34, 72]

max = reduce(lambda a, b: a if a > b else b, numbers)

print("The maximum number is:", max)

````

## Question 12: Multiple Exceptions

- Handle `ValueError` if the input is not a number.
- Handle `TypeError` if the input is not a string/integer.
- Ensure the program prints a friendly error message for both cases.

````python
def square_num(value):
    try:
        n = int(value)
        return n * n
    except ValueError:
        return "Error: Input is not a  number"
    except TypeError:
        return "Error: Input must be a string or an integer"


print(square_num("10"))     # 100
print(square_num(78))       # 6084
print(square_num("aayushma"))   # Error: Input is not a number
print(square_num(None))    # Error: Input must be a string or an integer
````

## Question 15: OOP - Inheritance
Create a parent class `Employee` with attributes `name` and `salary`.
- Create a child class `Manager` that inherits from `Employee` and adds an attribute `department`.
- Write a method in `Manager` to display the name, salary, and department.

````python
# Write your answer here
# Definition: Inheritance allows a class (child/subclass) to reuse attributes and methods of another class (parent/superclass).
# Purpose: Code reusability, avoid duplication.

class Employee:
    def __init__(Aayushma, name, salary): # using _init_ -> Parent attributes initialized,  Child attributes initialized
         Aayushma.name = name
         Aayushma.salary = salary

class Manager(Employee):
    def __init__(Aayushma, name, salary, department):
        super().__init__(name, salary)  # call parent constructor
        Aayushma.department = department

    def display_info(Aayushma):
        print("Name:", Aayushma.name)
        print("Salary:", Aayushma.salary)
        print("Department:",Aayushma.department)

# Create object of child class
manager1 = Manager("Aayushma", 100000 , "IT")

# Call method
manager1.display_info()
````

## Question 16: OOP - Encapsulation
Create a class `BankAccount` with a private attribute `_balance`.
- Implement methods `deposit(amount)` and `withdraw(amount)` to modify the balance.
- Ensure that `withdraw` does not allow taking out more money than the current balance.
- Create a `get_balance()` method to safely access the private balance.

````python
# Write your answer here
# Definition: Encapsulation hides the internal details of a class and provides controlled access to its data.
# Purpose: Protect data from accidental modification, enforce rules.

class BankAccount:
    def __init__(Aayushma, initial_balance=0):
        Aayushma.__balance = initial_balance 
        
# private attribute using name mangling->Name mangling in Python is a way the language protects class-private variables from being accessed accidentally from outside the class
#Aayushma refers to the current object.
#Aayushma.__balance → safe private variable inside the class.
#__balance is the private attribute belonging to that object.


    def deposit(Aayushma, amount):
        if amount > 0:
            Aayushma.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Deposit amount must be positive")

    def withdraw(Aayushma, amount):
        if amount > Aayushma.__balance:
            print("Insufficient balance")
        elif amount <= 0:
            print("Withdraw amount must be positive")
        else:
            Aayushma.__balance -= amount
            print(f"Withdrew: {amount}")

    def get_balance(Aayushma):
        return Aayushma.__balance  # safe access to private attribute



account = BankAccount(100000)
account.deposit(10000)        
account.withdraw(20000) 
account.withdraw(200000)  
    
print("Current Balance:", account.get_balance())  
````

## Question 17: OOP - Polymorphism
Create two classes, `Cat` and `Dog`.
- Both classes should have a method `speak()`.
- `Cat.speak()` should print "Meow".
- `Dog.speak()` should print "Woof".
- Write a function `animal_sound(animal)` that accepts an object and calls its `speak()` method, demonstrating polymorphism.

````python
# Write your answer here
# Definition: Polymorphism allows different classes to be used interchangeably if they implement the same method or interface.
# Purpose: Write flexible, reusable code that works with different objects.

class Cat:
    def speak(Aayushma):
        print("Meow")


class Dog:
    def speak(Aayushma):
        print("Woof")

# Function  to polymorphism
def animal_sound(animal):
    animal.speak()  # calls the speak() method of the object passed


cat = Cat()
dog = Dog()
# animal_sound() is a function that accepts any object and calls its speak() method

animal_sound(cat) 
animal_sound(dog)  
````

## Question 18: Static Methods
Create a class `MathUtils` with a static method `is_even(n)`.
- The method should return `True` if `n` is even, and `False` otherwise.
- Call this method without creating an instance of the class.

````python
# Write your answer here
# Static Method:
# A method that belongs to the class, not to any object.
# Does not use self.
# Can be called directly on the class.
# Why use @staticmethod?
# For utility functions that don’t need access to instance data.
# Keeps related functions grouped under the class.
# Static methods are like normal functions, but namespaced inside a class for organization.

class MathUtils:
    @staticmethod
    def is_even(n):
        return n % 2 == 0

# Calling static method without creating an object
print(MathUtils.is_even(1000))
print(MathUtils.is_even(19))   
````

## Question 19: Variable Arguments (*args and **kwargs)
Write a function `student_report` that accepts:
- A mandatory argument `name`.
- Arbitrary positional arguments (`*args`) for subject scores.
- Arbitrary keyword arguments (`**kwargs`) for additional details (like `city`, `age`).
- The function should print the name, calculate the average of the scores, and print the additional details.

````python
# Write your answer here
# Variable Arguments (*args and **kwargs)
# Write a function student_report that accepts:
# args → captures arbitrary positional arguments as a tuple.
# Here, 99, 94, 95 are captured in args.
# Useful for passing any number of scores.
# **kwargs → captures arbitrary keyword arguments as a dictionary.
# Here, city="Surkhet" and age=21 are captured in kwargs.


def student_report(name, *args, **kwargs):
    print(f"Name: {name}")
#     f"" → the f before the quotes indicates an f-string.
# {name} → Python replaces this with the current value of the variable name.
# So if name = "Aayushma", this line prints:
# Name: Aayushma
    
    if args:
        average_score = sum(args) / len(args)
        print(f"Average Score: {average_score}")
    else:
        print("No scores")
    
    if kwargs:
        print("Additional Details:")
        for key, value in kwargs.items():
            print(f"{key}: {value}")


student_report("Aayushma", 99, 94, 95, city="Surkhet", age=21)
````

## Question 20: Comprehensive System
Design a simple "Library Management System" using OOP and Exception Handling.
- Create a class `Library` with a list of available books.
- Add methods to `borrow_book(book_name)` and `return_book(book_name)`.
- If a user tries to borrow a book that isn't in the list, raise a custom exception `BookNotFoundError` (or use a standard `ValueError`).
- Ensure the state of the library updates correctly after each transaction.

````python
# Write your answer here
# Custom Exception
class BookNotFoundError(Exception):
    pass
    
# BookNotFoundError is a custom exception class that inherits from Python’s built-in Exception.
# pass means “do nothing”, because we don’t need to add any extra code or methods for this exception.

class Library:
    def __init__(Aayushma, books=None):
        
#Why books=None is used
# Default mutable arguments are risky:
# If you wrote def __init__(self, books=[]), the same list would be shared across all instances of the class.
# This can lead to unexpected behavior when one instance modifies the list.
# Using None avoids this problem:
# books=None is immutable, safe as a default.
        
        if books is None:
            books = []
        Aayushma.books = books  # list of available books

    def borrow_book(Aayushma, book_name):
        try:
            if book_name in Aayushma.books:
                   Aayushma.books.remove(book_name)
                   print(f"You have borrowed '{book_name}'")
            else:
                raise BookNotFoundError(f"'{book_name}' is not available in the library")
        except BookNotFoundError as e:
            print(e)

    def return_book( Aayushma, book_name):
        Aayushma.books.append(book_name)
        print(f"You have returned '{book_name}'")

    def show_books( Aayushma):
        if  Aayushma.books:
            print("Available books:", ", ".join( Aayushma.books))
        else:
            print("No books available in the library")



library = Library(["Python", "Javascrpit", "SQL"])

library.show_books()
library.borrow_book("Python")  # Borrow a book
library.borrow_book("C++")  # Book not available
library.show_books()
library.return_book("Python")  # Return book
library.show_books()
````
